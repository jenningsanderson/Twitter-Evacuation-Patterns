```ruby
#Connect to the Database
require 'mongo'
require 'json'
Mongo::Logger.logger.level = Logger::WARN #Suppress default debug messages
db = Mongo::Client.new("mongodb://epic-analytics.cs.colorado.edu:27017", {database: 'hurricane_sandy'})
coll = db['tweets']
```

```ruby
results = coll.find({:coordinates=> {'$ne'=>nil}, :created_at => {'$lt'=>Time.utc(2012,10,30)}}).distinct('user.screen_name').projection({coordinates:1,_id:0})
```

```ruby
output = {type: "FeatureCollection", features: []}
results.each_with_index do |res, idx|
  output[:features] << {type: "Feature", properties: {}, geometry: res["coordinates"]}
  if idx%100==0
    print "."
    print res
  end
end

File.open('all_geo_keyword.geojson', 'wb') do |file|
  file.write( output.to_json )
end
  
```
